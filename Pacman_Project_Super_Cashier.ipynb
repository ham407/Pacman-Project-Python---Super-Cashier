```python
from tabulate import tabulate
```

```python
class Transaction():
  """kelas transaksi untuk membuat objek dari class transaksi."""
  def __init__(self):
    self.items = []

  # method add_item untuk menambahkan barang
  def add_item(self):
    """menambahkan jumlah barang yang akan diinput."""
    n_items = int(input("\nMasukkan jumlah barang yang akan ditambahkan: "))
    if n_items >= 5:
      print("tdk bisa")
    else:
      for n in range(n_items):
        print(f"\n============Item ke-{n + 1}==========")
        name = str(input("Masukkan nama barang\t: "))
        qty = int(input("Masukkan jumlah barang\t: "))
        price = float(input("Masukkan harga barang\t: "))

        self.items.append([name,qty,price])
      # ("aqua", 2, 12000)

  # method update_item_name untuk mengupdate nama barang dari nama sebelumnya menjadi nama baru
  def update_item_name(self):
    """
    merubah nama barang jika terdapat kesalahan input.

    Parameter:
        old.name(str): nama barang sebelumnya
        new.name(str): nama barang baru


      break:
        untuk menghentikan literasi ketika nilai sudah sesuai
    """
    try:
      old_name = input("Masukkan nama barang yang ingin dirubah: ")
      new_name = str(input("Masukkan nama barang baru: "))


      for item in self.items:
        if item [0] == old_name:
          item[0] = new_name

          break # untuk menghentikan literasi ketika nilai sudah sesuai
    except ValueError:
      print("Input nama barang yang diberikan tidak valid. Masukkan tipe data string")


 # method update_item_qty untuk mengupdate jumlah barang yang akan dibeli jika terjadi perubahan
  def update_item_qty(self):
    """
    mengubah jumlah barang yang akan dibeli dengan menggunakan index

    Parameter:
      item_name = nama barang
      new_qty = jumlah barang baru yang akan dibeli

    Break:
    menghentikan proses literasi ketika nilai sudah sesuai

    """

    try:
      name = input("Masukkan nama barang yang dirubah: ")
      new_qty =int(input("Masukkan jumlah barang yang baru: "))

      for item in self.items:
       if item[0] == name:
        item[1] = new_qty # mendefinisikan tipe data

        # self.total_price = sum(item[1] * item[2] for item in self.items)
        break # untuk menghentikan literasi ketika nilai sudah sesuai
      print("Update item berhasil")

    except ValueError:
       print("Input nilai yang diberikan tidak valid.Masukkan nilai bilangan bulat")



  # method update_item_price untuk mengupdate harga barang yang akan dibeli
  def update_item_price(self):
    """
    merubah harga barang yang baru dengan menggunakan index

    Parameter:
    item_name = nama barang yang akan dibeli
    new_price = harga barang yang baru

    """
    try:
      item_name = input("Masukkan nama barang")
      new_price = int(input("Masukkan harga barang yang ingin dirubah"))

      for item in self.items:
        if item[0] == item_name:
          item[2] = new_price

    except ValueError:
      print("Input nilai yang diberikan tidak valid. Masukkan nilai bilangan")


        # self.total_price = sum(item[1] * item[2] for item in self.items)
      #item[2]

  # method delete_item untuk menghapus barang yang akan dibeli
  def delete_item(self):
    """
    menghapus barang belanja yang tidak dibeli

    Parameter:
    nama_item(str) = nama barang yang akan dihapus


    Return:
    None
    """
    try:
      nama_item =str(input("Masukkan nama barang yang dihapus: "))


      for item in self.items:
        if item[0] == nama_item:
          self.items.remove(item)
        break # untuk menghentikan literasi ketika nilai sudah sesuai

    except ValueError:
      print("Input nilai yang diberikan tidak valid. Masukkan nilai bilangan")


  # method reset_transaction untuk mengreset transaksi yang tidak jadi dibeli
  def reset_transaction(self):

    """menghapus semua jenis transaksi dengan menggunakan list kosong."""

    self.items = [] # untuk mengosongkan nilai dalam list self.items

  # method check_order untuk mengecek barang yang akan dibeli
  def check_order(self):
    """
    mengecek daftar belanja apakah sudah sesuai atau belum

    return:
      mengembalikan nilai barang
    """

    if not self.items:
      print("Terdapat kesalahan input data")
      return # untuk mengembalikan nilai

    for item in self.items:
      if "" in item:
        print("Terdapat kesalahan input data")
        return # untuk mengembalikan nilai

    print("Pemesanan sudah benar")

  # method total_price untuk menghitung total harga
  def total_price(self):
    """menghitung total harga barang untuk menentukan apakah mendapatkan diskon atau tidak."""

    total_transaksi = sum(item[1] * item[2] for item in self.items)
    diskon = 0 #sebagai nilai awal karena belum tau diskon yang akan diberikan

    if total_transaksi > 500000:
      diskon = 0.10
      print(f"Total belanja Rp. {total_transaksi} mendap1at diskon sebesar 10%: Rp. {total_transaksi * diskon}")

    elif total_transaksi > 300000:
      diskon = 0.08
      print(f"Total belanja Rp. {total_transaksi} mendapat diskon sebesar 10%: Rp. {total_transaksi * diskon}")

    elif total_transaksi > 200000:
      diskon = 0.05
      print(f"Total belanja Rp. {total_transaksi} mendapat diskon sebesar 10%: Rp. {total_transaksi * diskon}")

    # mengurangi total transaksi dengan diskon
    total_transaksi -= total_transaksi * diskon
    return total_transaksi

  # method untuk menampilkan tabel
  def tampil(self):
    """menampilkan tabel nama items,qty, harga per item dan total harga"""

    headers = ["nama items","qty","harga per item","total harga"]

    tabel_data =[]

    for item in self.items:
      total_harga = item[1] * item[2]
      tabel_data.append([item[0], item[1], item[2], total_harga])


    print("daftar harga supermarket")
    print("") # sebagai spasi
    print(tabulate(tabel_data,headers, tablefmt='grid', stralign="center"))

    total_belanja = self.total_price()
    print("Total belanjaan Anda adalah: Rp.", total_belanja)


  #method menampilkan display menu
  def display_menu(self):
    """menampilkan display menu"""

    print("\nmenu")
    print("1. tambah barang")
    print("2. update nama barang")
    print("3. update jumlah barang")
    print("4. update harga barang")
    print("5. hapus barang")
    print("6. reset saat transaksi")
    print("7. mengecek transaksi")
    print("8. total harga")
    print("9. tampilkan belanjaan")
    print("0. keluar")
    print(" ")

  # method memulai tampilan
  def start(self):
    """memulai pilihan pada menu"""

    try:
      while True:
        self.display_menu()
        pilihan = int(input("Masukkan pilihan anda: "))

        if pilihan == 1:
          self.add_item()
        elif pilihan == 2:
          self.update_item_name()
        elif pilihan == 3:
          self.update_item_qty()
        elif pilihan == 4:
          self.update_item_price()
        elif pilihan == 5:
          self.delete_item()
        elif pilihan == 6:
          self.reset_transaction()
        elif pilihan == 7:
          self.check_order()
        elif pilihan == 8:
          self.total_price()
        elif pilihan == 9:
          self.tampil()
        elif pilihan == 0:
          print(f"Terimakasih telah berbelanja ditempat kami")
          break
        else:
          print("input harus 0-9")

    except ValueError:
      print("Input harus menggunakan angka")


```
