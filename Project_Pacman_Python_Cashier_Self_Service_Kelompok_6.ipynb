## **Codelab**

```python
from tabulate import tabulate

class Transaction:
  """
  Kelas yang merepresentasikan suatu transaksi pembelanjaan.

  Atribut:
  - items (list): Lis yang berisi daftar item dalam bentuk list dengan format [name_item, jumlah_item, harga_per_item].
  """

  def __init__(self):
    """
    Inisialisasi objek Transaction().

    Parameters:
    Tidak menggunakan parameter

    Returns:
    Objek Transaction baru.
    """
    self.items = []

  def add_item(self):
    """
    Menambahkan item baru kedalam transaksi dalam bentuk list.

    Paramters:
    - name (str): Nama item yang akan ditambahkan
    - qty (int): Jumlah item yang akan ditambahkan
    - price (float): Harga per item

    Returns:
    Tidek mengembalikan nilai
    """

    n_items = int(input("Masukkan jumlah item yang ingin ditambahkkan: "))

    # menggunakan perulangan untuk memberikan input item sesuai jumlah items
    for i in range(n_items):
      print(f"\n============Item ke-{i+1}============")
      name = input("Masukan nama item\t: ")
      qty = int(input("Masukkan jumlah item\t: "))
      price = float(input("Masukkan harga per item\t: "))

      # menyimpan data input kedalam list `items` menggunakan append
      self.items.append([name, qty, price])

  def update_item_name(self):
    """
    Mengupdate nama item yang sudah dalam list transaksi.

    Parameters:
    - name(str): Nama item yang akan diupdate namanya.
    - old_name(str): Item item yang baru

    Returns:
    Tidak mengembalikan nilai
    """

    old_name = input("Masukan nama barang yang ingin dirubah: ")
    new_name = input("Masukkan nama barang yang baru: ")

    # trnsct_123.update_item_name("Aqua", "Air Mineral")

    for i in self.items:
      if i[0] == old_name:
        i[0] = new_name

        # `break` digunakan untuk menghentikan iterasi ketika nilai sudah ditemukan/sesuai.
        print("\nUpdate item berhasil")
        break

  def update_item_qty(self): # menggunakan parameter `name` sebagai penentu barang karena tidak ada parameter `id barang`
    """
    Mengupdate jumlah item untuk item yang sudah ada di dalam list transaksi.

    Parameters:
    - name (str): Nama item yang akan diupdate jumlahnya.
    - new_qty (int): Jumlah item yang baru.

    Returns:
    Tidak mengembalikan nilai
    """

    try:
      name = input("Masukan nama barang yang ingin dirubah: ")
      new_qty = int(input("Masukkan jumlah barang yang baru: "))

      for i in self.items:
        if i[0] == name:
          i[1] = int(new_qty) # mendefinisikan tipe data

          # `break` digunakan untuk menghentikan iterasi ketika nilai sudah ditemukan/sesuai.
          break
      print("Update item berhasil")

    except ValueError:
      print("Input yang diberikan tidak valid. Masukkan bilangan bulat")

  def update_item_price(self):
    """
    Mengupdate harga item untuk item yang sudah ada di dalam list transaksi.

    Parameters:
    - name (str): Nama item yang akan diupdate harganya.
    - new_price (int): Harga item yang baru.

    Returns:
    Tidak mengembalikan nilai
    """

    try:
      name = input("Masukan nama barang yang ingin dirubah: ")
      new_price = float(input("Masukkan harga barang yang baru: "))

      for i in self.items:
        if i[0] == name:
          i[2] = float(new_price) # mendefinisikan tipe data

          # `break` digunakan untuk menghentikan iterasi ketika nilai sudah ditemukan/sesuai.
          break
      print("Update item berhasil")

    except ValueError:
      print("Input yang diberikan tidak valid. Masukkan bilangan bulat")

  def delete_item(self):
    """
    Menghapus item untuk item yang sudah ada di dalam list transaksi.

    Parameters:
    - name (str): Nama item yang akan dihapus keberadaannya.

    Returns:
    Tidak mengembalikan nilai
    """
    name = input("Masukan nama barang yang ingin dihapus: ")

    for i in self.items:
      if i[0] == name:
        self.items.remove(i)

        # `break` digunakan untuk menghentikan iterasi ketika nilai sudah ditemukan/sesuai.
        break
    print("Delete item berhasil")

  def reset_transaction(self):
    """
    Menghapus semua item transaksi dan mengembalikan keadaan transaksi ke kondisi awal.

    Parameters:
    Tidak memerlukan parameter.

    Returns:
    Tidak mengembalikan nilai.
    """
    self.items = []     #self.items diberikan list kosong `[]` untuk mengosongkan nilai

  def check_order(self):
    """Memeriksa apakah transaksi sudah benar atau terdapat kesalahan input data.

    Parameters:
    Tidak memerlukan parameter.

    Returns:
    Tidak mengembalikan nilai.
    """

    # memeriksa apakah `self.items` saat ini kosong? Jika kosong, tampilkan output yang di return
    if not self.items:
      print("Terdapat kesalahan input data")
      return

    # jika daftar item tidak kosong, maka jalan kondisi berikut
    for item in self.items:
      if '' in item:
        print("Terdapat kesalahan input")
        return

    print("Pemesanan sudah benar")

  def total_price(self):
    """
    Menghitung total belanjaan sebelum diskon dan menampilkan perkiraan diskon yang didapat.

    Parameters:
    Tidak memerlukan parameter.

    Returns:
    float: Total harga belanjaan setelah diskon
    """
    total = sum(item[1] * item[2] for item in self.items) # menggunakan loop untuk mengiterasi setiap item yang ada dalam daftar `self_items`/ daftar transaksi
    discount = 0

    # gunakan percabangan untuk mengecek diskon berdasarkan total transaksi
    if total > 500000:
      discount = 0.10
      print(f"Anda mendapatkan diskon 10% sebesar\t: Rp. {total*discount}")
    elif total > 300000:
      discount = 0.08
      print(f"Anda mendapatkan diskon 8% sebesar\t: Rp. {total*discount}")
    elif total > 200000:
      discount = 0.05
      print(f"Anda mendapatkan diskon 5% sebesar\t: Rp. {total*discount}")

    # mengurangi total belanja dengan diskon
    total -= total*discount
    return total

  def display_transaction(self):
    """
    Menampilkan detail transaksi belanja dalam bentuk tabel/tabular.

    Parameters:
    Tidak memerlukan parameter.

    Returns:
    Tidak mengembalikan nilai.
    """
    # mendefinisikan header tabel
    headers = ["Nama Item", "Jumlah Item", "Harga per Item", "Total Harga"]

    # variabel table_data nantinya berisi nilai total_belanja, diawal definisikan dengan list kosong
    table_data = []

    for item in self.items:
      # self.items merupakan index yang berisi ["nama", qty, price]
      total_harga = item[1] * item[2]
      table_data.append([item[0], item[1], f"Rp. {item[2]}", f"Rp. {total_harga}"])

    # memanggil method `total_price()` dan menyimpannya dalam `total_belanja`
    total_belanja = self.total_price()
    table_data.append(["", "", "Jumlah : ",f"Rp. {total_belanja}"])

    # tampilan tabel dengan menggunakan library `tabulate`
    print("\n",tabulate(table_data, headers=headers, tablefmt="grid"))

  # membuat method menghitung uang kembalian
  def calculate_change(self):
    """
    Menghitung jumlah kembalian yang didapat customer berdasarkan total belanja dan jumlah uang yang dibayarkan.

    Parameters:
    Tidak memerlukan parameter. Customer diminta untuk memasukkan jumlah yang yang dibayarkan.

    Returns:
    Jika kembalian diberikan, method akan mencetak kembalian dan mengembalikan nilainya.
    Jika jumlah uang yang dibayarkan tidak mencukupi, method akan mencetak pesanan kesalahan dan mengembalikan None.
    """

    total_belanja = self.total_price()
    print(f"Total belanja anda saat ini adalah\t: Rp. {total_belanja}")

    # memberikan input jumlah uang yang dibayarkan
    amount_paid = float(input("Masukkan jumlah uang yang dibayarkan: Rp. "))

    # menghitung kembalian
    if amount_paid < total_belanja:
      return "Uang yang dibayarkan tidak mencukupi"
    else:
      change = amount_paid - total_belanja
      print(f"Kembalian anda: Rp. {change}")
      return change


  def display_menu(self):
    print("\nMenu:")
    print("1. Tambah Item")
    print("2. Ubah Nama Item")
    print("3. Ubah Jumlah Item")
    print("4. Ubah Harga Item")
    print("5. Hapus Item")
    print("6. Reset Transaksi")
    print("7. Cek Transaksi")
    print("8. Tampilkan Total Belanjan")
    print("9. Pembayaran")
    print("0. Keluar")
    print(" ")

  def start(self):
    try:

      while True:
        self.display_menu()
        choice = int(input("Pilih menu(0-9): "))

        if choice == 1:
          self.add_item()
        elif choice == 2:
          self.update_item_name()
        elif choice == 3:
          self.update_item_qty()
        elif choice == 4:
          self.update_item_price()
        elif choice == 5:
          self.delete_item()
        elif choice == 6:
          self.reset_transaction()
        elif choice == 7:
          self.check_order()
        elif choice == 8:
          self.display_transaction()
        elif choice == 9:
          self.calculate_change()
        elif choice == 0:
          print("Terimakasih telah menggunakan layanan kami. Sampai jumpa!")
          break
        else:
          print("Masukkan tidak valid. Silahkan pilih angka antara 0-9")

    except ValueError:
      print("Input yang diberikan tidak valid")
```
